```python
import os
import numpy as np
import matplotlib.pyplot as plt
from src.metrics import psnr
import cv2
```

```python
from src.data.common import image2kspace, generate_mask, kspace2image, crop_image
```

```python
def generate_mask(shape, center_fraction, ratio):
    '''
    apply the mask to k-space data that the low frequency is located in corner. 
    '''
    assert len(shape) == 2
    num_cols = shape[-1]
    num_low_freqs = int(round(num_cols * center_fraction))
    prob = (num_cols * ratio - num_low_freqs) / (num_cols - num_low_freqs)
    mask = np.random.rand(1, num_cols) < prob
    pad_loc = (num_cols - num_low_freqs + 1) // 2
    mask[:, pad_loc:pad_loc+num_low_freqs] = True
    return mask
```

```python
data_path = '/home/rzhou/ssd_cache/fastMRI_sampled_npy/singlecoil_train/'
files = os.listdir(data_path)
image = np.load(os.path.join(data_path, files[15]))
image = np.abs(image)
# image = crop_image(image, (160, 480), (0, 320))
kspace = image2kspace(image)
mask = generate_mask(kspace.shape, 0.3, 0.5)
image_zerofilled = kspace2image(kspace*mask + 0.0)
image_zerofilled = np.abs(image_zerofilled)
print(psnr(image, image_zerofilled))
cv2.imwrite('zf.png', image_zerofilled*255.)
cv2.imwrite('ori.png', image*255.)
```
